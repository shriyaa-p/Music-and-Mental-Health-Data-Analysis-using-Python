```python
# loading the dataset in a pandas dataframe named 'df'
df = pd.read_csv(r'/content/drive/My Drive/projects/mxmh_dataset.csv')
```

```python
# top 5 rows of the dataset
print('top 5 rows of the dataset:\n',df.head(5))

# bottom 5 rows of the dataset
print('\nbottom 5 rows of the dataset:\n',df.tail(5))
```

```python
# size of dataset and data type of columns
print('number of rows and columns in the dataset (size of the dataset): ',df.shape)
print('\ncolumns and their data types:\n',df.dtypes)
```

```python
# keeping only required columns
required_df = df[['Age', 'Primary streaming service', 'Hours per day', 'Frequency [Classical]', 'Frequency [Country]', 'Frequency [EDM]', 'Frequency [Folk]', 'Frequency [Gospel]', 'Frequency [Hip hop]', 'Frequency [Jazz]', 'Frequency [K pop]', 'Frequency [Latin]', 'Frequency [Lofi]', 'Frequency [Metal]', 'Frequency [Pop]', 'Frequency [R&B]', 'Frequency [Rap]', 'Frequency [Rock]', 'Frequency [Video game music]', 'Anxiety', 'Depression', 'Insomnia', 'OCD', 'Music effects']]
print('required dataset columns and column types:\n',required_df.dtypes)
print('size of the required dataset: ',required_df.shape)
```

```python
# sum of numeric columns
for cols in required_df.select_dtypes(include='number'):
    print('sum of ',cols,': ',required_df[cols].sum())
```

```python
# average of numerical columns :
try:
  for cols in required_df.select_dtypes(include='number'):
    print('average of ',cols,': ',required_df[cols].sum()/len(required_df[cols]))
except TypeError:
  print('error')
```

```python
# converting all string values to lowercase
for i in df.columns:
  if df[i].dtype==object:
    df[i]=df[i].str.lower()

# counting group count for - 'Primary streaming service'
print('count for each type of',required_df.groupby(['Primary streaming service'])['Primary streaming service'].count())
```

```python
# summary stats
print(required_df.describe())
```

```python
# correlation table for anxiety, depression, insomnia and ocd with respect to age
numerical_columns = required_df.select_dtypes(include='number').columns
numerical_columns = numerical_columns.drop('Age')
print(required_df[numerical_columns].corr())
```

```python
# checking duplicates
print('number of duplicate row entries: ',required_df.duplicated().sum())
print('number of duplicate column entries: ',required_df.columns.duplicated().sum())
```

```python
# counting null values
null_count = required_df.isnull().sum()
print('null values for each column:\n',null_count)
print('\ntotal null values:',null_count.sum())
print('\npercentage of null values: ',null_count.sum()*100/len(required_df))
```
